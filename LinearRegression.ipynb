## 1. Linear Regression
### 1.1. Fitting a Straight Line
The most frequently used curve-fitting criterion is the least-squares criterion. The problem is to determine the parameters of the function $y=f(x)$ to minimize the sum
$$S=\sum_{i=1}^n|y_i-f(x_i)|^2,$$
of $n$ data points, $(x_i, y_i)$ where $i=1, 2, \dots, n$.

Suppose a model of the form $y = ax + b$ is expected and it has been decided to use the $n$ data points $(x_i, y_i)$ for $i = 1, 2,\dots , n$ to estimate $a$ and $b$. Then,
 $$S=\sum_{i=1}^n|y_i-(ax_i+b)|^2=\sum_{i=1}^n|y_i-ax_i-b|^2.$$
The necessary condition for optimality is that the partial derivatives $\dfrac{\partial S}{\partial a}$ and $\dfrac{\partial S}{\partial b}$ equal to zero. That is,
\begin{align}
\dfrac{\partial S}{\partial a}&=\dfrac{\partial }{\partial a}\left[\sum_{i=1}^n\left(y_i-ax_i-b\right)^2\right],\nonumber\\
0&=2\sum_{i=1}^n\left(y_i-ax_i-b\right)(-x_i),\nonumber\\
0&=\sum_{i=1}^n(y_i-ax_i-b)(x_i).\\
\dfrac{\partial S}{\partial b}&=\dfrac{\partial }{\partial b}\left[\sum_{i=1}^n\left(y_i-ax_i-b\right)^2\right],\nonumber\\
0&=2\sum_{i=1}^n\left(y_i-ax_i-b\right)(-1),\nonumber\\
0&=\sum_{i=1}^n(y_i-ax_i-b).\\
\end{align}
From equation (1) we get
\begin{align*}
\sum_{i=1}^n(y_i-ax_i-b)(x_i)&=0,\nonumber\\
\sum_{i=1}^nx_iy_i-a\sum_{i=1}^nx^2_i-b\sum_{i=1}^nx_i&=0,\nonumber\\
a\sum_{i=1}^nx^2_i+b\sum_{i=1}^nx_i&=\sum_{i=1}^nx_iy_i.
\end{align*}
Also, from equation (2) we have
\begin{align*}
\sum_{i=1}^n(y_i-ax_i-b)&=0,\nonumber\\
\sum_{i=1}^ny_i-a\sum_{i=1}^nx_i-\sum_{i=1}^nb&=0,\nonumber\\
a\sum_{i=1}^nx_i+b\cdot n&=\sum_{i=1}^ny_i.
\end{align*}
The preceding equations can be used to solve for $a$ and $b$ once all the values for $x_i$ and $y_i$ are substituted into them. Then, we obtain
\begin{align*}
a&=\dfrac{n\sum x_iy_i-\sum x_i\sum y_i}{n\sum x^2_i-(\sum x_i)^2}\\
b&=\dfrac{n\sum x^2_iy_i-\sum x_iy_i\sum x_i}{n\sum x^2_i-(\sum x_i)^2}
\end{align*}
where $a$ and $b$ are the slope and the intercept of the linear regression, respectively.

### 1.2. Fitting a Power Curve
We use the least-squares criterion to fit a curve of the form $y = ax^m$, where $m$ is fixed, to a given collection of data points. Applying the criterion on the curve requires minimization of
$$S=\sum_{i=1}^n|y_i-f(x_i)|^2=\sum_{i=1}^n|y_i-ax^m_i|^2.$$
A necessary condition for optimality is that the derivative $\dfrac{dS}{da}=0$, which yields
$$\dfrac{dS}{da}=2\sum_{i=1}^n\left(y_i-ax_i^m\right)(-x_i^m)=0$$
Solving the equation for $a$ yields
\begin{align*}
2\sum_{i=1}^n\left(y_i-ax_i^m\right)(-x_i^m)&=0\\
\sum_{i=1}^n\left(x_i^my_i-ax_i^mx_i^m\right)&=0\\
\sum_{i=1}^nx_i^my_i-a\sum_{i=1}^n\left(x_i^m\right)^2&=0\\
a=\dfrac{\sum x_i^my_i}{\sum x_i^{2m}}
\end{align*}

## 2. Linear Models
### 2.1. Linear Regression
Linear regression is a supervised learning model used for prediction and for explaining the relationship between a dependent and an independent variable.

Ordinary least squares (OLS) is the most commonly used method for fitting the linear regression model. The main purpose of a model is to use some data in order to predict future observations. In addition, regularization techniques such as lasso and ridge regression overcome some of the problems of OLS. 

Ridge regression shrinks the coefficients estimates towards zero, which improves the variability, reduces the variance while slightly increases the bias and raises the overall accuracy of the model. When the number of predictors is high, ridge regression results in models that are difficult to interpret.

Lasso regression improves both accuracy and model interpretability by selecting which coefficients to shrink and by shrinking  some of them exactly to zero.

### 2.2. Ordinary Least Squares
#### 2.2.1. Estimation of Ordinary Least Squares
Given a $n$-tuples training data $(x_{1j}, \dots, x_{nj}, y_i)$, a multiple regression model describing the relationship between the dependent variable, $y$ and the independent variables $x_j$ is
$$y=\beta_0+\sum_{j=1}^{N}\beta_jx_j+\varepsilon,$$
where $\varepsilon$ is the error term and $\beta_0$ is the intercept. We predict the value of $y$ based on the value of $x_i$, by computing the following equation
$$\hat{y}=\hat{\beta}_0+\sum_{j=1}^N\hat{\beta}_jx_j,$$
where the hat symbols indicate the estimated coefficients. In order to fit this mode, we typically use a method called \textit{least squares} or \textbf{\textit{Ordinary Least Squares}} (OLS). The idea behind OLS is to find the parameters $\hat{\beta}_j$ such that the regression line is the closest line to the data points $(x_{nj}, y)$. The OLS method estimates the coefficients $\beta_1, \dots, \beta_j$ by minimizing a quantity known as the \textbf{\textit{residual sum of squares}} (RSS). This is defined as
$$RSS=\sum_{i=1}^N (y_i-\hat{y})^2=\sum_{i=1}^N\left[y_i-\left(\hat{\beta}_0+\sum_{j=1}^N\hat{\beta}_jx_j\right)\right]^2,$$
where $y_i$ is the actual value and $\hat{y}$ is the predicted value. Differentiate the equation above with respect to $\hat{\beta}_i$ and solving for $\hat{\beta}_i$ yields the OLS coefficients, which we denote $\hat{\beta}_{OLS}$. Hence, the OLS estimates are
$$\hat{\beta}_{OLS}=\underset{\beta}{\arg\min}\sum_{i=1}^N\left[y_i-\left(\beta_0+\sum_{j=1}^N\beta_jx_j\right)\right]^2$$

#### 2.2.2. Derivation of Multiple Regression
The ordinary least squares estimator is obtained by minimizing the objective function 
\begin{align*}
	RSS&=\sum_{i=1}^N (y_i-\hat{y})^2=\sum_{i=1}^N\left[y_i-\left(\hat{\beta}_0+\sum_{j=1}^p\hat{\beta}_jx_{ji}\right)\right]^2
\end{align*}
Using the first-order conditions we have
\begin{align*}
	\dfrac{\partial RSS}{\partial \hat{\beta}_0} &= -2\sum_{i=1}^N\left[y_i-\left(\hat{\beta}_0+\sum_{j=1}^p\hat{\beta}_jx_{ji}\right)\right]=0,\\
	\dfrac{\partial RSS}{\partial \hat{\beta}_j} &= -2\sum_{i=1}^N\left[y_i-\left(\hat{\beta}_0+\sum_{j=1}^p\hat{\beta}_jx_{ji}\right)\right]x_{ji}=0.\\
\end{align*}
Then,
\begin{align*}
	\sum_{i=1}^Ny_i-\sum_{i=1}^N\hat{\beta}_0 - \sum_{i=1}^N\sum_{j=1}^p\hat{\beta}_jx_{pi}&=0,\\
	\sum_{i=1}^Ny_i-N\hat{\beta}_0 - \sum_{i=1}^N\sum_{j=1}^p\hat{\beta}_jx_{pi}&=0,\\
	N\hat{\beta}_0+\sum_{i=1}^N\sum_{j=1}^p\hat{\beta}_jx_{pi}&= \sum_{i=1}^Ny_i. 
\end{align*}
And,
\begin{align*}
	\sum_{i=1}^N\left[y_i-\left(\hat{\beta}_0+\sum_{j=1}^p\hat{\beta}_jx_{ji}\right)\right]x_{ji}&=0,\\
	\sum_{i=1}^Nx_{ji}y_i-\hat{\beta}_0\sum_{i=1}^Nx_{ji}-\sum_{j=1}^p\hat{\beta}_jx^2_{ji}&=0,\\
	\hat{\beta}_0\sum_{i=1}^Nx_{ji}+\sum_{j=1}^p\hat{\beta}_jx^2_{ji}&=\sum_{i=1}^Nx_{ji}y_i
\end{align*}

#### 2.2.3. Matrix-Vector Form
Consider the general  linear regression model
\begin{align*}
	y_i&=\beta_0 + \beta_1x_{1i}+\beta_2x_{2i}+\cdots+\beta_{p}x_{pi}+\varepsilon_{i},\\
	y_i &= \beta_0 + \sum_{j=1}^p\beta_{j}x_{ji} +\varepsilon_i
\end{align*}
where $i=1, 2, \dots, N$, and $j=1, 2,\dots, p$. This can be written in terms of system of equations as 
\begin{align*}
	y_1 &= \beta_0 + \beta_1x_{11}+\beta_2x_{21}+\cdots+\beta_{p}x_{p1}+\varepsilon_{1}\\
	y_2 &= \beta_0 + \beta_1x_{12}+\beta_2x_{22}+\cdots+\beta_{p}x_{p2}+\varepsilon_{2}\\
	&\vdots\\
	y_N &= \beta_0 + \beta_1x_{1N}+\beta_2x_{2N}+\cdots+\beta_{p}x_{pN}+\varepsilon_{N}
\end{align*}
The system can be written in matrix form as. 
\begin{align*}
	\begin{bmatrix}
		y_1\\ y_2\\ \vdots\\ y_N
	\end{bmatrix}=
	\begin{bmatrix}
		1 & x_{11} & x_{21} & \cdots & x_{p1} \\
		1 & x_{12} & x_{22} & \cdots & x_{p2}\\
		\vdots & \vdots& \vdots & \ddots & \vdots\\
		1 & x_{1N} & x_{2N} & \cdots & x_{pN}\\
	\end{bmatrix}
	\begin{bmatrix}
		\beta_0\\ \beta_1 \\ \beta_2 \\ \vdots \\ \beta_p
	\end{bmatrix} + 
	\begin{bmatrix}
		\varepsilon_{1} \\ \varepsilon_{2} \\ \vdots \\ \varepsilon_{N}
	\end{bmatrix}
\end{align*}

\subsubsection{Matrix-Vector Form}
The sum of squared error, SSE, is defined as 
$$SSE=\sum_{i=1}^N\left(y_i-\hat{y}_i\right)^2$$
The mean squared error, MSE, is defined as 
$$MSE=\dfrac{1}{N}\sum_{i=1}^N\left(y_i-\hat{y}_i\right)^2$$
Then,
\begin{align*}
	SSE&=\sum_{i=1}^N\left(y_i-\hat{y}_i\right)^2\\
	&=\sum_{i=1}^N\left(\boldsymbol{\beta}^T\boldsymbol{x}_i-y_i\right)^2\\
	&=\left(\boldsymbol{X}\boldsymbol{\beta}-\boldsymbol{y}\right)^T\left(\boldsymbol{X}\boldsymbol{\beta}-\boldsymbol{y}\right)\\
	&=\|\boldsymbol{X}\boldsymbol{\beta}-\boldsymbol{y}\|^2\\
	&=\|\boldsymbol{\hat{y}}-\boldsymbol{y}\|^2
\end{align*}
In ordinary least squared regression, we want to minimize the residual squared error (SSE). This minimization problem involves computing the derivative of SSE with respect to $\boldsymbol{\beta}$. In other words, we compute the gradient of SSE as
\begin{align*}
	\nabla SSE(\boldsymbol{\beta})&=\dfrac{\partial }{\partial\boldsymbol{\beta}}SSE(\boldsymbol{\beta})\\
	&=\dfrac{\partial }{\partial\boldsymbol{\beta}}\left(\boldsymbol{X}\boldsymbol{\beta}-\boldsymbol{y}\right)^T\left(\boldsymbol{X}\boldsymbol{\beta}-\boldsymbol{y}\right)\\
	&=\dfrac{\partial }{\partial\boldsymbol{\beta}}\left[\boldsymbol{\beta}^T\boldsymbol{X}^T\boldsymbol{X}\boldsymbol{\beta}-\boldsymbol{\beta}^T\boldsymbol{X}^T\boldsymbol{y}-\boldsymbol{y}^T\boldsymbol{X}\boldsymbol{\beta}-\boldsymbol{y}^T\boldsymbol{y}\right]\\
	&=\dfrac{\partial }{\partial\boldsymbol{\beta}}\left[\boldsymbol{\beta}^T\boldsymbol{X}^T\boldsymbol{X}\boldsymbol{\beta}-2\boldsymbol{\beta}^T\boldsymbol{X}^T\boldsymbol{y}-\boldsymbol{y}^T\boldsymbol{y}\right]\\
	&=2\boldsymbol{X}^T\boldsymbol{X}\boldsymbol{\beta}-2\boldsymbol{X}^T\boldsymbol{y}
\end{align*}
Setting the derivative of SSE to zero we obtain the following normal equations
$$\boldsymbol{X}^T\boldsymbol{X}\boldsymbol{\beta}=\boldsymbol{X}^T\boldsymbol{y}$$
Then, the vector of regression coefficients $\boldsymbol{\beta}$ is $\boldsymbol{\beta}=\left(\boldsymbol{X}^T\boldsymbol{X}\right)^{-1}\boldsymbol{X}^T\boldsymbol{y}$. In addition, the expression for the predicted outcomes can be computed as 
$$\boldsymbol{\hat{y}}=\boldsymbol{X}\boldsymbol{\beta}=\boldsymbol{X}\left(\boldsymbol{X}^T\boldsymbol{X}\right)^{-1}\boldsymbol{X}^T\boldsymbol{y}$$
The matrix $\boldsymbol{X}\left(\boldsymbol{X}^T\boldsymbol{X}\right)^{-1}\boldsymbol{X}^T\boldsymbol{y}$ is an orthogonal projection. 

## Example of Linear Regression Using Scikit-Learn

```python
from sklearn.datasets import load_diabetes # load a dataset
from sklearn.model_selection import train_test_split # used to split the data set to training set and test set
from sklearn.linear_model import LinearRegression # linear regression model
from sklearn.metrics import mean_squared_error, r2_score # used to measure model performance
import matplotlib.pyplot as plt

X, y = load_diabetes(return_X_y=True)
X = X[:, [2]] # use only one feature
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=20,shuffle=False)

# fit linear regression model on  the training data. By default, an intercept is added to the model.
# we can control this behavior by setting the fit_intercept parameter
regressor = LinearRegression().fit(X_train, y_train)

# model evaluation
y_pred = regressor.predict(X_test)
print(f"Mean squared error: {mean_squared_error(y_test, y_pred):.2f}")
print(f"Coefficient of determination: {r2_score(y_test, y_pred):.2f}")

print("\n")
# plotting the results
fig, ax = plt.subplots(ncols=2, figsize=(10, 5), sharex=True, sharey=True)
ax[0].scatter (X_train, y_train, label="Train data points")
ax[0].plot(
    X_train,
    regressor.predict(X_train),
    linewidth=3,
    color="orange",
    label="Model prediction"
)
ax[0].set(xlabel="Feature", ylabel="Target", title="Train set")
ax[0].legend()

ax[1].scatter(X_test, y_test, label="Test data points")
ax[1].plot(
    X_test,
    y_pred,
    linewidth=3,
    color="orange",
    label="Model predictions"
)
ax[1].set(xlabel="Feature", ylabel="Target", title="Test set")
ax[1].legend()
plt.show()
```

```python
import mglearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt

X, y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=20, random_state=42)
lr = LinearRegression().fit(X_train, y_train)
y_pred = lr.predict(X_test)

print(f"linear regression coefficient: {lr.coef_}")
print(f"linear regression intercept: {lr.intercept_}")

print(f"Training set score: {r2_score(X_train, y_train): .2f}")
print(f"Test set score: {r2_score(X_test, y_test): .2f}")
print(f"Mean squared error: {mean_squared_error(y_test, y_pred): .2f}")

fig, ax = plt.subplots(ncols=2, figsize=(10,5))
ax[0].scatter(X_train, y_train, label="Train data points")
ax[0].plot(
    X_train,
    lr.predict(X_train),
    linewidth=3,
    color="orange",
    label="Model predictions"
)
ax[0].legend()

ax[1].scatter(X_test, y_test, label="Test data points")
ax[1].plot(
    X_test,
    lr.predict(X_test),
    linewidth=3,
    color="orange",
    label="Model predictions"
)
ax[1].legend()
plt.show()
```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# import data
housing = fetch_california_housing()
X = housing.data
y = housing.target

# split the dataset into two different sets: training and testing (evaluation)
# we use the 20% of the original dataset for evaluation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# normalize data using StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# instance a random forest and we train it with the training dataset
regr = RandomForestRegressor(max_depth=2, random_state=0)
regr.fit(X_train, y_train)

# predict over evaluation dataset and show MSE of the trained model
y_pred = regr.predict(X_test)
print(f"Mean squared error: {mean_squared_error(y_pred, y_test): .2f}")
```
